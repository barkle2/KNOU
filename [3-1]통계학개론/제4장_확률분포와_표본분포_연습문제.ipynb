### 02. 취직시험이 있었다. 점수가 450과 500 사이인 사람은 그 직업에 적당하다고 한다. 만일 평균점수가 400이고, 표준편차라 50인 정규분포를 이룬다면 그 직업에 적당하다고 생각되는 점수는 몇 %에 해당되는가?

```python
from scipy.stats import norm

# 평균과 표준편차
mean = 400
std_dev = 50

# 적당한 점수 범위
lower_bound = 450
upper_bound = 500

# 450점 이하의 누적확률
prob_below_450 = norm.cdf(lower_bound, mean, std_dev)
# 500점 이하의 누적확률
prob_below_500 = norm.cdf(upper_bound, mean, std_dev)

# 450~500점 사이의 확률
prob_between_450_500 = prob_below_500 - prob_below_450

# 백분율로 변환
percentage = prob_between_450_500 * 100

print(f"적당한 점수(450~500점)에 해당되는 비율: {percentage:.2f}%")
```
